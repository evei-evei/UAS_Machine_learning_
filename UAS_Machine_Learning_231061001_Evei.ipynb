```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Baris ini membaca file Excel dengan nama ipm.xlsx yang berada di folder /konten/, lalu menyimpannya ke dalam variabel df sebagai sebuah DataFrame.

```python
df = pd.read_excel("/content/ipm.xlsx", engine='openpyxl')
X = df.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Sekolah')
plt.xlabel('Rata-rata lama sekolah')
plt.ylabel('Harapan lama sekolah')
plt.show()
```

pada awalnya, peningkatan rata-rata lama sekolah berhubungan dengan penurunan yang signifikan pada harapan lama sekolah. Hal ini mungkin menujukkan bahwa pada kondisi tertentu, faktor-faktor seperti akses pendidikan atau tingkat pendidikan dasar yang lebih tinggi menyebabkan perbedaan dalam harapan lama sekolah.
setelah rata-rata lama sekolah mencapai angka tertentu (sekitar 6-10 tahun). penurunan harapan lama sekolah tidak lagi terlalu signifikan. ini mungkin menujukkan bahwa di tingkat rata-rata lama sekolah yang lebih tinggi, harapan lama sekolah mencapai stabilitas atau saturasi.

Kode berikut digunakan untuk mengimplementasikan algoritma K-Means clustering dengan jumlah klaster yang sudah ditentukan, lalu mengelompokkan data ke dalam klaster.

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

Kode di atas digunakan untuk memvisualisasikan hasil klasterisasi menggunakan scatter plot. Grafik ini menunjukkan data yang dikelompokkan ke dalam klaster berdasarkan hasil algoritma K-Means.

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Rata-rata lama sekolah (Tahun)' )
plt.ylabel('Harapan lama sekolah (Tahun')
plt.legend()
plt.show()
```

Gambar tersebut menunjukkan hasil klasterisasi berdasarkan rata-rata lama sekolah dan harapan lama sekolah, yang terbagi menjadi lima klaster dengan karakteristik berbeda. Klaster dengan rata-rata lama sekolah dan harapan lama sekolah tinggi cenderung mencerminkan kelompok yang memiliki akses pendidikan lebih baik, sementara klaster dengan nilai lebih rendah menunjukkan keterbatasan dalam pendidikan. Hubungan antara kedua variabel ini terlihat positif secara umum, meskipun terdapat variasi dalam beberapa klaster. Klasterisasi ini dapat membantu memahami pola distribusi pendidikan untuk perencanaan kebijakan yang lebih efektif.

```python
Kmeans1 = KMeans(n_clusters = 1, init = 'k-means++', random_state = 42)
y_kmeans1 = Kmeans1.fit_predict(X)
```

Kode di atas digunakan untuk memvisualisasikan hasil klasterisasi dengan algoritma K-Means pada data yang memiliki dua fitur utama: %BPJS dan %Sanitasi. Grafik yang dihasilkan membantu memahami bagaimana data dikelompokkan ke dalam klaster dan di mana centroid klaster berada.

```python
plt.scatter (X[y_kmeans1 == 0, 0], X[y_kmeans1 == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(Kmeans1.cluster_centers_[:, 0], Kmeans1.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('%BPJS')
plt.ylabel('%Sanitasi')
plt.legend()
plt.show()
```

Gambar tersebut menunjukkan grafik sebaran data pelanggan yang dikelompokkan menjadi satu klaster (Cluster 1) berdasarkan persentase BPJS (%BPJS) dan persentase sanitasi (%Sanitasi).  Titik-titik merah mewakili data pelanggan individual, sementara titik kuning mewakili centroid dari Cluster 1.  Sebaran data menunjukkan bahwa sebagian besar pelanggan dalam Cluster 1 memiliki persentase BPJS dan sanitasi yang relatif dekat satu sama lain.  Analisis lebih lanjut diperlukan untuk menentukan karakteristik spesifik dari Cluster 1 dan implikasinya.
